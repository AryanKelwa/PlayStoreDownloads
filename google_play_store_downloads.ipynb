```python
def convert_data(x):
    if str(x)[-1]=='M':
        res=x.replace('M','')
        res=float(res)*1000
        return float(res)
    elif str(x)[-1]=='k':
        res=str(x).replace('k','')
        return float(res)



df['Size']=df['Size'].apply(lambda x:convert_data(x))

```
